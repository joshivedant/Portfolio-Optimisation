Importing Libraries

```python
import numpy as np
import pandas as pd
import pandas_ta as ta
import matplotlib.pyplot as plt
```

Import Data

```python
listname = pd.read_csv("_ind_nifty100list.csv")
NTPC = pd.read_csv("NTPC.csv")
NTPC.head()
```

Pre-process

```python
def momentum(df, day):
    return df.Close - df.Close.shift(day)

def RSI(df, period):
    return df.ta.rsi(close='Close', length=period)

def psar(df):
    return df.ta.psar(high=df['High'], low=df['Low'], close=df['Close'], af0=0, af=0, max_af=0)

def atr(df, period):
    return df.ta.atr(high=df['High'], low=df['Low'], close=df['Close'], length=period)

def tr(df):
    return df.ta.true_range(high=df['High'], low=df['Low'], close=df['Close'])

def chaikin(df):
    return df.ta.cmf(high=df['High'], low=df['Low'], close=df['Close'], volume=df['Volume'])

def obv(df):
    return df.ta.obv(close=df['Close'], volume=df['Volume'])
```

```python
NTPC['r1'] = np.log(NTPC.Close / NTPC.Close.shift(1))
NTPC['r2'] = np.log(NTPC.Close / NTPC.Close.shift(2))
NTPC['r3'] = np.log(NTPC.Close / NTPC.Close.shift(3))
NTPC['r4'] = np.log(NTPC.Close / NTPC.Close.shift(4))

NTPC['r5'] = np.log(NTPC.High / NTPC.Open)
NTPC['r6'] = np.log(NTPC.High / NTPC.Open.shift(1))
NTPC['r7'] = np.log(NTPC.High / NTPC.Open.shift(2))
NTPC['r8'] = np.log(NTPC.High / NTPC.Open.shift(3))

NTPC['r9'] = np.log(NTPC.High.shift(1) / NTPC.Open.shift(1))
NTPC['r10'] = np.log(NTPC.High.shift(2) / NTPC.Open.shift(2))
NTPC['r11'] = np.log(NTPC.High.shift(3) / NTPC.Open.shift(3))

NTPC['r12'] = np.log(NTPC.Low / NTPC.Open)
NTPC['r13'] = np.log(NTPC.Low / NTPC.Open.shift(1))
NTPC['r14'] = np.log(NTPC.Low / NTPC.Open.shift(2))
NTPC['r15'] = np.log(NTPC.Low / NTPC.Open.shift(3))

NTPC['Momentum'] = momentum(NTPC, 10)
NTPC['RSI_14'] = RSI(NTPC, 14)
# temp = psar(NTPC)
# print(temp)
NTPC['atr'] = atr(NTPC, 14)
NTPC['tr'] = tr(NTPC)
NTPC['chaikin'] = chaikin(NTPC)
NTPC['obv'] = obv(NTPC)

NTPC.loc[(NTPC.Close.shift(-1) - NTPC.Close) / NTPC.Close > 0, 'output'] = 1
NTPC.loc[(NTPC.Close.shift(-1) - NTPC.Close) / NTPC.Close <= 0, 'output'] = 0
```

```python
NTPC.dropna(inplace=True)
#NTPC.drop(['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume'], axis=1, inplace=True)
NTPC.set_index('Date', inplace=True)

NTPC.head()
```

Train-test Split

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(NTPC, test_size=0.2)

x_train = train[NTPC.columns[6:27]]
y_train = train[NTPC.columns[27]]

x_test = test[NTPC.columns[6:27]]
y_test = test[NTPC.columns[27]]
```

SVM

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

model = SVC().fit(x_train, y_train)

print(accuracy_score(y_train, model.predict(x_train)))

print(accuracy_score(y_test, model.predict(x_test)))
```

```python
NTPC['signal'] = model.predict(NTPC[NTPC.columns[6:27]])
```
